```python
num = 5

# if 문 : bool 자료형
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 나머지는 F

score = 99

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
fruits = ['딸기','사과','바나나','배','수박']

# fruits 리스트에 딸기가 포함되어 있다면 딸기가 포함됨 없다면 딸기가 없음
# in, if


if "딸기" in fruits:
    print("딸기가 포함됨")
else:
    print("딸기가 없음")
```

```python
# and : x and y (x,y 가 모두 참인 경우 참)
# or : x or y (x,y 가 둘 중 하나만 참이면 참)
# not x :

# score 가 90이상이고, grade 가 A 인 경우 합격 나머지 불합격
score, grade = 85, 'A'

if score>=90 and grade == "A":
    print("합격")
else:
    print("불합격")
```

```python
# score 가 90이상이거나, grade 가 A 인 경우 합격 나머지 불합격

if score>=90 or grade == "A":
    print("합격")
else:
    print("불합격")
```

```python
# 나이가 10살 이상이고, 키가 140cm 이상이면 "입장 가능합니다" 그렇지 않으면 "입장할 수 없습니다"

age = 12
height = 145

if age>=10 and height>=140:
    print("입장 가능합니다")
else:
    print("입장할 수 없습니다")
```

```python
# 5세 미만 "무료" , 5세 이상 ~ 18세 미만 "청소년 할인 대상", 18세 이상 ~ 65세 미만 "일반 요금", 65세 이상 "경로 우대 할인 대상"

age = 20

if age < 5:
    print("무료")
elif age >= 5 and age < 18:
    print("청소년 할인 대상")
elif age >= 18 and age < 65:
    print("일반 요금")
else:
    print("경로 우대 할인 대상")
```
